```python
import numpy as np 
import matplotlib.pyplot as plt 
import random
```

```python
def tb(t,j,L):               #t hopping, j disorder, L size
    
    h = np.zeros((L, L))
    for i in range(L-1):
        h[i,i+1] = t
        h[i+1,i] = t
    
    
    diagonal = np.zeros(L)

    for i in range(len(diagonal)):
        diagonal[i]= j*random.uniform(-0.5, 0.5)
    diag = np.diag(diagonal)

    h = h + diag
    
    return h
                           
```

```python
L=500
eigvals,eigvecs = np.linalg.eig(tb(1,1,L))
```

```python
sites = np.arange(0,L,1)
plt.plot(sites, eigvals[np.argsort(eigvals)], '.')
```
