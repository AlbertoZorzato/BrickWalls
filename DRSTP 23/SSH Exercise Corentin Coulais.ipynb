```python
import numpy as np 
import matplotlib.pyplot as plt 
import random
```

```python



def hamiltonian(t1,t2):
    
    h = np.array([[0,t1,0,0,0,0],
                 [t1,0,t2,0,0,0],
                 [0,t2,0,t1,0,0],
                 [0,0,t1,0,t2,0],
                 [0,0,0,t2,0,t1],
                 [0,0,0,0,t1,0]])
    
    return h
                           
```

```python
def tb(t1):
    
    h = np.array([[0,t1,0,0,0,0,0],
                 [t1,0,t1,0,0,0,0],
                 [0,t1,0,t1,0,0,0],
                 [0,0,t1,0,t1,0,0],
                 [0,0,0,t1,0,t1,0],
                 [0,0,0,0,t1,0,t1],
                 [0,0,0,0,0,t1,0]])
    
    return h
                           
```

```python
def nhmetamaterial(a,b,c,d):
    
    metamat = np.array([[0,b,0,0,0,0,d],
                        [a,0,c,0,0,0,0],
                        [0,d,0,b,0,0,0],
                        [0,0,a,0,c,0,0],
                        [0,0,0,d,0,b,0],
                        [0,0,0,0,a,0,c],
                        [b,0,0,0,0,d,0]])
    
    return metamat
                           
```

```python
t1=1
A = []
list = np.arange(-5,5,0.1)
sites = np.arange(0,6,1)
for a,t2 in enumerate(list):
    eigvals, eigvecs = np.linalg.eig(hamiltonian(t1,t2))
    A.append(eigvals)
   
A = np.asarray(A)

plt.plot(list,A,'x')

    
```

```python
t1 = 1
t2 =4

eigvals, eigvecs = np.linalg.eig(hamiltonian(t1,t2))
plt.plot(sites, eigvals[np.argsort(eigvals)],'.')
    
```
