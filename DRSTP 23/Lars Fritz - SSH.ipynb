```python
import numpy as np 
import matplotlib.pyplot as plt 
import random
```

```python
def tb(t1,t2,j,d,L):               #t1,t2 hopping, j asymmetry, d disorder, L size
    
    h = np.zeros((L, L), dtype=complex)
    for i in range(L-1):
        if i % 2 == 0:
            h[i, i + 1] = t1
            h[i + 1, i] = t1
        else:
            h[i, i + 1] = t2 - 1j * j
            h[i + 1, i] = t2 + 1j * j
        
       
    
    
    diagonal = np.zeros(L)

    for i in range(len(diagonal)):
        diagonal[i]= d*random.uniform(-0.5, 0.5)
    diag = np.diag(diagonal)

    h = h + diag
    
    return h
                           
```

```python
tb(1,2,1,1,10)
```

```python
t1=1; t2=1; j=1; d=0; L=200


L=600
eigvals,eigvecs = np.linalg.eigh(tb(t1,t2,j,d,L))
```

```python
sites = np.arange(0,L,1)
plt.plot(sites, eigvals[np.argsort(eigvals)], '.')
```

```python
plt.plot(sites,np.abs(eigvecs)**2)
```

```python
t2=1; j=0.5; d=0; L=500
list = np.arange(0,2,0.1)
A=np.zeros((len(list),L))
for z, t1 in enumerate(list):
    
    
    eigvals,_ = np.linalg.eigh(tb(t1,t2,j,d,L))
    
    A[z,:] = eigvals
    
    
```
