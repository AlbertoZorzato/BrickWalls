```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
rb= .005
rd= .00005
people = 0
timerange = np.arange(0,1000000,1)
A = []

for time, t in enumerate(timerange):
    
    for i in range(people):
        csi = np.random.rand()
        if csi<rd:
            people-=1
    
    csi = np.random.rand()
    if csi<rb:
        people+=1
    #if people < 0:
    #    break
    A.append(people)
    
A = np.asarray(A)
```

```python
plt.ylim(0,max(A))
plt.xlabel('time')
plt.ylabel('people')
plt.plot(timerange, A,'-', c='m')
```
